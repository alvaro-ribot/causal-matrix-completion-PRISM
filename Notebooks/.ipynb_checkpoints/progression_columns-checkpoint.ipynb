```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.colors as colors
import json

from scipy.stats import iqr

import seaborn as sns
```

```python
algos = ['mean_over_cells', 'mean_over_drugs', 'fixed_effect',
        'squared_si_cells', 'squared_csi_cells', 'squared_csi_drugs','squared_si_fe',
        'nnm', 'nnm_fe']
```

```python
# w/o mean_over_drugs
algos = ['mean_over_cells', 'fixed_effect',
        'squared_si_cells', 'squared_csi_cells', 'squared_csi_drugs','squared_si_fe',
        'nnm', 'nnm_fe']
```

```python
dfr2 = pd.DataFrame(columns = ['algorithm', 'rows&columns', 'experiment', 'R2'])

for alg in algos:
    
    R2 = json.load(open('../R2_'+alg+'.txt'))
    
    
    if 'squared' in alg: alg = alg[8:]

    for column in range(5, 280, 5):
        values = R2[str(column)]
        row = pd.DataFrame({'algorithm': [alg]*20,
              'rows&columns': [column]*20,
              'experiment': [i for i in range(20)],
              'R2': values})
        dfr2 = pd.concat([dfr2, row], ignore_index = True)
```

```python
# for low data regime:


dfr2 = pd.DataFrame(columns = ['algorithm', 'rows&columns', 'experiment', 'R2'])

for alg in algos:
    
    R2 = json.load(open('../performance_lowdata/R2_low_'+alg+'.txt'))
    
    
    if 'squared' in alg: alg = alg[8:]

    for column in range(1, 31):
        values = R2[str(column)]
        row = pd.DataFrame({'algorithm': [alg]*20,
              'rows&columns': [column]*20,
              'experiment': [i for i in range(20)],
              'R2': values})
        dfr2 = pd.concat([dfr2, row], ignore_index = True)
```

```python
sns.set_theme()
#sns.reset_orig()
```

```python
def showlines(data, title = '', errorbar = False, metric = 'R2'):
    if not errorbar: ax = sns.lineplot(data=data, x='rows&columns', y=metric, hue='algorithm', errorbar=None)
    else: ax = sns.lineplot(data=data, x='rows&columns', y=metric, hue='algorithm')
    #ax.set(xlabel='number of rows&columns oberved per missing entry', title = title)
    ax.set(xlabel='obs', ylim = (-0.1,1))
    plt.title(title, fontsize=30)
    plt.show()
```

```python
def rename(df, before, after):
    df.loc[df["algorithm"] == before, "algorithm"] = after
    return df
```

```python
algos = ['mean_over_cells', 'mean_over_drugs', 'fixed_effect',
        'si_cells', 'csi_cells', 'csi_drugs','si_fe',
        'nnm', 'nnm_fe']

new = ['mean-over-contexts', 'mean-over-actions', 'FE',
        'SI', 'SI-mean-contexts', 'SI-mean-actions','SI-FE',
        'NNM', 'NNM-FE']
```

```python
for b, a in zip(algos, new):
    dfr2 = rename(dfr2, b, a)
```

```python
sns.set(font_scale=1)
showlines(dfr2, 'All', True)
```

```python
dfmse = pd.DataFrame(columns = ['algorithm', 'rows&columns', 'experiment', 'MSE'])

for alg in algos[:6]:
    
    MSE = json.load(open('../MSE_'+alg+'.txt'))

    for column in range(5, 280, 5):
        values = MSE[str(column)]
        row = pd.DataFrame({'algorithm': [alg]*20,
              'rows&columns': [column]*20,
              'experiment': [i for i in range(20)],
              'MSE': values})
        dfmse = pd.concat([dfmse, row], ignore_index = True)
```

```python
sns.set_theme()
ax = sns.lineplot(data=dfmse, x='rows&columns', y='MSE', hue='algorithm')
ax.set(xlabel='number of rows&columns oberved per missing entry')
plt.show()
```
